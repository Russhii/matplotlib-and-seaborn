```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

# Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel(" Rush")
plt.ylabel(" Rushikesh")
plt.title(1)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(2)

plt.subplot(2,2,2)
plt.plot(x**2,y**2)
plt.subplot(2,2,3)
plt.plot(y,x)
```

# objects oriented way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) # these values always between 0 and 1
axis1.plot(x,y) 

axis2 = fig.add_axes([0.4,0.4,0.3,0.3])
axis2.plot(x,y)
```

# Types of plots

# saving plots


```python
fig =plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
